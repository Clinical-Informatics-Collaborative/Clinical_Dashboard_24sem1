```python
import pandas as pd
import random
import geopandas as gpd
random.seed(1256)

condition_df1 = pd.read_csv('data/Synthetic data/project1/conditions.csv')
medication_df1 = pd.read_csv('data/Synthetic data/project1/medications.csv')
patient_df1 = pd.read_csv('data/modified data/project1/patients.csv')

condition_df2 = pd.read_csv('data/Synthetic data/project2and3/conditions.csv')
medication_df2 = pd.read_csv('data/Synthetic data/project2and3/medications.csv')
patient_df2 = pd.read_csv('data/modified data/project2/patients.csv')
```

```python
patient_df = pd.concat([patient_df1, patient_df2], ignore_index=True)
medication_df = pd.concat([medication_df1, medication_df2], ignore_index=True)
condition_df = pd.concat([condition_df1, condition_df2], ignore_index=True)
```

```python
# Delete rows where DESCRIPTION is not "Ischemic heart disease (disorder)" / "Diabetes mellitus type 2 (disorder)"
condition_df = condition_df[condition_df['DESCRIPTION'] == 'Ischemic heart disease (disorder)']
condition_df
```

```python
# Filter patient_df based on condition_df's 'PATIENT' column
patient_df = patient_df[patient_df['Id'].isin(condition_df['PATIENT'])]

#reasign address
melbourne_suburbs =[
  "Carlton", "Carlton North", "Docklands", "East Melbourne",
  "Flemington", "Kensington", "Melbourne", "North Melbourne",
  "Parkville", "Port Melbourne", "Southbank", "South Wharf",
  "South Yarra", "West Melbourne", "Albert Park", "Balaclava",
  "Elwood", "Middle Park", "Ripponlea", "St Kilda", "St Kilda East",
  "St Kilda West", "South Melbourne", "Abbotsford", "Alphington",
  "Burnley", "Clifton Hill", "Collingwood", "Cremorne", "Fairfield",
  "Fitzroy", "Fitzroy North", "Princes Hill", "Richmond"
]
patient_df["COUNTY"] = random.choices(melbourne_suburbs, k=len(patient_df))



patient_df
```

```python
# Filter medication_df based on condition_df's 'PATIENT' column
medication_df = medication_df[medication_df['ENCOUNTER'].isin(condition_df['ENCOUNTER'])]
medication_df = medication_df[medication_df['PATIENT'].isin(patient_df['Id'])]

medication_df
```

```python
patient_df.to_csv('data/heart disease in melbourne/patients.csv', index=False)
medication_df.to_csv('data/heart disease in melbourne/medications.csv', index=False)
condition_df.to_csv('data/heart disease in melbourne/conditions.csv', index=False)
```
